```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/github-ranking-recent.csv')
df["repo"] = df["repo_name"] + '/' + df["username"]
```

```python
df = df[df["item"] != 'top-100-forks']
```

```python
df.drop(['item', 'rank', 'repo_url', 'description', 'username', 'repo_name'], axis=1, inplace=True)
df.drop_duplicates(inplace = True)
df = df[['repo', 'language', 'stars', 'forks', 'watchers', 'pullRequests', 'milestones',
       'issues', 'totalIssues', 'created', 'last_commit']]
```

```python
df.corr()
```

```python
df_sum = df.groupby(['language']).sum().sort_values(by = ['forks'], ascending = [False])
```

```python
df_sum.to_csv('../Data/sum_lang.csv')
```
